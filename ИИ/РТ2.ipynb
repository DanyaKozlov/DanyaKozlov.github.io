### задача 1.3.4

```python
import numpy as np
import random
from random import randint
b = random.randint(3,15)
d = random.randint(3,15)
c = d-1;
a = np.array([0])
a = np.resize(a,(b,d))
for i in range(b):
    if i == 0 or i == b-1:
        a[i]=1;
    for k in range(d):
        if k==0 or k == c:
            a[i][k]=1;
print(a)
```

### задача 1.3.6

```python
import numpy as np
import random
from random import randint
c = random.randint(1,10)
d = random.randint(1,10)
a = np.array([0])
a = np.resize(a,(c,d))

for i in range(c):
    for k in range(d):
        b = random.randint(0,40)
        a[i][k] = b;
print(a)
print(np.shape(a))
print(a.size)
print(a.ndim)
```

### задание 3.3.2

```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/iris.csv')
df.head()


feat_length = np.array(df['sepal_length_cm']) 
feat_width = np.array(df['sepal_width_cm'])

minmax_scale = preprocessing.MinMaxScaler(feature_range = (0, 1))
z_scale = preprocessing.StandardScaler()

scaled_feat_length = minmax_scale.fit_transform(feat_length.reshape(-1, 1))
scaled_feat_width = z_scale.fit_transform(feat_width.reshape(-1, 1))

print(scaled_feat_length[0:5], '\n'*2, scaled_feat_width[0:5])
```
